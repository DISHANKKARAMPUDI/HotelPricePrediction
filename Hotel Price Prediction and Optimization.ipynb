## 5. Feature Engineering

Combining the weekend and weekday stay columns into one feature simplifies analysis of the total duration guests stay.

## 6. Classifying Records as per seasons of hotel booking

Grouping months into seasons can reveal seasonal trends, which might influence occupancy rates and pricing strategies.


```python
def get_season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    else:
        return 'Fall'

df_main['season'] = df_main['arrival_date_month'].apply(get_season)
```

## 7. Removing records which are not informative

Bookings with no adults, children, or babies are likely invalid, so removing them ensures that all bookings have at least one occupant.

```python
# Remove rows where adults, children, and babies are all 0
df_main= df_main[
    (df_main['adults'] > 0) | (df_main['children'] > 0) | (df_main['babies'] > 0)
]
```

Rows where the number of adults is 0 might represent invalid or suspicious bookings, as hotel rooms are usually booked for at least one adult. Removing such records helps ensure that the dataset reflects realistic scenarios.

```python
df_main = df_main[df_main['adults'] > 0]
```

This prevents the model from learning from or being influenced by unrealistic or potentially erroneous data points, leading to more accurate and reliable predictions.


## 8. Standardizing 'meal' column for clarification

Standardizing the meal column ensures consistency in categories and avoids errors during analysis or grouping by meal type.

```python
df_main['meal'] = df_main['meal'].str.strip().str.lower().replace({
    'bb': 'bed & breakfast',
    'hb': 'half-board',
    'fb': 'full-board',
    'sc': 'self-catering'
})
```

## Encoding the columns

```python
from sklearn.preprocessing import OneHotEncoder

# One-Hot Encoding for the 'hotel' column
hotel_onehot = pd.get_dummies(df_main1['hotel'], prefix='hotel')

# Target Encoding for the 'arrival_date_month' column based on adr
# First, compute the mean adr for each month
month_target_encoding = df_main1.groupby('arrival_date_month')['adr'].mean()

# Map the month to its corresponding adr mean value
df_main1['arrival_date_month_encoded'] = df_main1['arrival_date_month'].map(month_target_encoding)

# Drop the original 'arrival_date_month' column
df_main1.drop(columns=['arrival_date_month'], inplace=True)

# Now concatenate the one-hot encoded 'hotel' columns to the main dataframe
df_main1 = pd.concat([df_main1, hotel_onehot], axis=1)

# Drop the original 'hotel' column
df_main1.drop(columns=['hotel'], inplace=True)
```

- One-Hot Encoding for 'hotel' column: The pd.get_dummies() function is used to perform one-hot encoding on the 'hotel' column. The prefix='hotel' ensures that the new columns generated have a 'hotel' prefix for easier identification.

- Target Encoding for 'arrival_date_month': The target encoding is done by calculating the mean adr (average daily rate) for each unique value in the 'arrival_date_month' column. The groupby() function is used to group by the 'arrival_date_month' and compute the mean.

- Mapping the Encoded Values:The map() function is used to map each month in the 'arrival_date_month' to its corresponding target encoding value.

- Dataframe Modifications:The original 'arrival_date_month' column is dropped after encoding, and the one-hot encoded 'hotel' columns are concatenated to the main dataframe. The drop() function ensures the original categorical columns are removed to avoid redundancy.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Initialize and train the Linear Regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Make predictions
y_pred_linear = linear_reg.predict(X_test)

# Evaluate the model
mae_linear = mean_absolute_error(y_test, y_pred_linear)
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)

# Print performance metrics
print("Linear Regression Performance:")
print(f"MAE: {mae_linear}")
print(f"MSE: {mse_linear}")
print(f"R² Score: {r2_linear}")

# Residual Plot
residuals = y_test - y_pred_linear
plt.figure(figsize=(10, 6))
plt.scatter(y_pred_linear, residuals, alpha=0.5, edgecolor='k')
plt.axhline(0, color='red', linestyle='--', label='Zero Residual')
plt.title("Residual Plot: Linear Regression")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.legend()
plt.grid(True)
plt.show()

# Histogram of Residuals
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, color='blue', alpha=0.7)
plt.title("Distribution of Residuals: Linear Regression")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

- Importing Linear Regression and Evaluation Metrics: LinearRegression is imported from sklearn.linear_model to initialize and train a linear regression model. mean_absolute_error, mean_squared_error, and r2_score are imported from sklearn.metrics to evaluate the model performance based on different metrics.

- Model Initialization and Training: The LinearRegression() object is created and stored in the variable linear_reg. The fit() function is used to train the linear regression model using X_train and y_train.

- Evaluating the Model: Three performance metrics are calculated to evaluate the model: mean_absolute_error (MAE) measures the average magnitude of the errors in the predictions. mean_squared_error (MSE) penalizes larger errors more than MAE. r2_score gives a measure of how well the regression predictions approximate the real data (a value of 1 means perfect prediction).

- Printing Model Performance: The model's performance is printed, showing the MAE, MSE, and R² score for the test set predictions, with an R² score indicating how much of the variance in the target variable is explained by the model (0.4736 in this case).

### Data Normalization

```python
from pyspark.sql.functions import col, min, max

# Min-max normalization for 'adr' (average daily rate)
stats = hotel_booking_df_cleaned.select(min("adr").alias("min_adr"), max("adr").alias("max_adr")).collect()
min_adr, max_adr = stats[0]['min_adr'], stats[0]['max_adr']

if max_adr == min_adr:
    print("Warning: adr values have no variance. Normalized column will be all zeros.")
    hotel_booking_df_cleaned = hotel_booking_df_cleaned.withColumn("normalized_adr", col("adr") - min_adr)
else:
    hotel_booking_df_cleaned = hotel_booking_df_cleaned.withColumn(
        "normalized_adr", (col("adr") - min_adr) / (max_adr - min_adr)
    )

hotel_booking_df_cleaned = hotel_booking_df_cleaned.fillna({"adr": 0})
```

```python
from pyspark.sql.functions import expr,when 

hotel_booking_df_cleaned = hotel_booking_df_cleaned.fillna({
    "stays_in_week_nights": 0,
    "stays_in_weekend_nights": 0,
    "total_of_special_requests": 0
})
# Add a new column 'total_stay_duration'
hotel_booking_df_cleaned = hotel_booking_df_cleaned.withColumn('total_stay_duration', expr('stays_in_weekend_nights + stays_in_week_nights'))
hotel_booking_df_cleaned.select('stays_in_weekend_nights', 'stays_in_week_nights', 'total_stay_duration').show(5)

# Add a derived column for "special requests per week"
hotel_booking_df_cleaned = hotel_booking_df_cleaned.withColumn(
    "special_requests_per_week",
    when(
        (col("stays_in_week_nights") + col("stays_in_weekend_nights")) > 0,
        col("total_of_special_requests") / (col("stays_in_week_nights") + col("stays_in_weekend_nights"))
    ).otherwise(0)
)
hotel_booking_df_cleaned.select("total_of_special_requests", "stays_in_week_nights", "stays_in_weekend_nights", "special_requests_per_week").show(5)

hotel_booking_df_cleaned.select("special_requests_per_week").describe().show()
```

```python
from pyspark.ml.feature import VectorAssembler, StandardScaler
import xgboost as xgb
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np

# Assemble features
assembler = VectorAssembler(inputCols=feature_columns, outputCol="features")
data_with_features = assembler.transform(hotel_booking_df_cleaned)

# Scale features
scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=True)
scaler_model = scaler.fit(data_with_features)
data_with_features = scaler_model.transform(data_with_features)

# Use scaled features for the model
final_data = data_with_features.select("scaled_features", "normalized_adr")

# Convert PySpark DataFrame to NumPy arrays (sample together to avoid mismatch)
sampled_data = final_data.rdd.map(lambda row: (row["scaled_features"], row["normalized_adr"])).sample(False, 0.1, seed=42).collect()
features_np = np.array([row[0] for row in sampled_data])
labels_np = np.array([row[1] for row in sampled_data])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features_np, labels_np, test_size=0.2, random_state=42)

# Train XGBoost Regressor
xgboost_regressor = xgb.XGBRegressor(
    max_depth=1,
    learning_rate=0.1,
    n_estimators=20,
    objective='reg:squarederror',
    random_state=3
)
xgboost_regressor.fit(X_train, y_train)

# Make predictions
y_pred = xgboost_regressor.predict(X_test)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("XGBoost Regressor Performance:")
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"R²: {r2}")

# Visualization: Predictions vs Actuals
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, edgecolor='k')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color="red", linestyle="--", label="Ideal Fit")
plt.title("XGBoost Regressor: Predictions vs Actual Values")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.legend()
plt.grid(True)
plt.show()

# Residual Plot
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, alpha=0.5, edgecolor='k')
plt.axhline(0, color="red", linestyle="--", label="Zero Residual")
plt.title("Residual Plot: XGBoost Regressor")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.legend()
plt.grid(True)
plt.show()

# Histogram of Residuals
plt.figure(figsize=(10, 6))
plt.hist(residuals, bins=30, color="green", alpha=0.7)
plt.title("Distribution of Residuals: XGBoost Regressor")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
